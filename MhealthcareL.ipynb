- AS Algorithms will only understand numerical values we cannot pass the data yet.
- We have to perform feature engineering on the features of the data 
- but before applying feature engineering, we have to understand what are the data types of inputs
- and split them into following groups
- *numbers*
- *categorical*
- *no transformation features*

```python
cat_df = X_train.select_dtypes(include= 'object')
cat_df
```

```python
df.columns
```

```python
df.head(2)
```

```python
num_df = X_train[['age','bmi','daily_steps','sleep_hours','stress_level','doctor_visits_per_year','hospital_admissions','medication_count','insurance_coverage_pct','previous_year_cost']]
num_df
```

```python
no_trans = X_train[['diabetes','hypertension','heart_disease','asthma']]
no_trans
```

```python
# noww transforming our cat_df and num_df using feature engineering
```

## Step 3 - Feature Engineering on inputs 

#### Performing One Hot Encoding(O.H.E) on categorical features(cat_df) of X_train 

```python
from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder(sparse_output=False).set_output(transform='pandas')
cat_df_trans = OHE.fit_transform(cat_df)
```

```python
cat_df_trans
```

#### Performing MinMaxScaler on numerical features(num_df) of X_train (Normalization)

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().set_output(transform='pandas')
num_df_trans = scaler.fit_transform(num_df)
```

```python
num_df_trans
```

```python
X_train_trans = pd.concat([cat_df_trans,num_df_trans,no_trans],axis=1)
```

```python
X_train_trans
```

## Step 4- Model training using Algorithmn X_test_trans and y_train

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
model.fit(X_train_trans,y_train)
```

## Step 5 - Feature engineering on Test data 

#### inorder to pass our test data to model
- first we have to transform the data by applying feature engineering

```python
cat_df = X_test.select_dtypes(include='object')
cat_df
```

```python
num_df = X_test[['age','bmi','daily_steps','sleep_hours','stress_level','doctor_visits_per_year','hospital_admissions','medication_count','insurance_coverage_pct','previous_year_cost']]
num_df
```

```python
no_trans = X_test[['diabetes','hypertension','heart_disease','asthma']]
no_trans
```

```python
cat_df_trans = OHE.transform(cat_df)
```

```python
num_df_trans = scaler.transform(num_df)
```

```python
cat_df_trans
```

```python
num_df_trans
```

```python
X_test_trans = pd.concat([cat_df_trans,num_df_trans,no_trans],axis=1)
```

```python
X_test_trans
```

## Step 6 - Prediction on X_test_trans

#### here we are pasing our test inputs (X_test_trans) to teh model to get our prediction output(y_pred) 

```python
y_pred = model.predict(X_test_trans)
```

```python
y_pred
```

## Step 7 Passing y-test(actual) and y_pred(model_prediction) to evaluation metrics

#### in this step we are passing our actual target variable and model prediction to evaluation metrics 
- To check how accurate our model is performing

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

this shows are model is performing 74% accurately
